```python
df["Date"] = pd.to_datetime(df["Date"]) 
df["Date"].sample(5)
```

```python
df['year'] = df["Date"].dt.year
df['month'] = df['Date'].values.astype('datetime64[M]')
```

```python
df['Longitude'].fillna(999, inplace = True)
df['Latitude'].fillna(999, inplace = True)
```

```python
def category(df_severity):
    if df_severity == 1:
        return 'Fatal'
    elif df_severity == 2:
        return 'Serious'
    elif df_severity == 3:
        return 'Slight'
    else:
        return 'Unknown'
```

```python
df['Accident_Category'] = df['Accident_Severity'].apply(category)
```

```python
df.index = pd.DatetimeIndex(df['Date'])
```

```python
plt.figure(figsize=(15,6))
plt.title('Distribution of accidents per day', fontsize=16)
plt.tick_params(labelsize=14)
sns.distplot(df.resample('D').size(), bins=60);
```

```python
df_daily = pd.DataFrame(df.resample('D').size())
df_daily['MEAN'] = df.resample('D').size().mean()
df_daily['STD'] = df.resample('D').size().std()
UCL = df_daily['MEAN'] + 3 * df_daily['STD']
LCL = df_daily['MEAN'] - 3 * df_daily['STD']

plt.figure(figsize=(15,6))
df.resample('D').size().plot(label='Accidents per day')
UCL.plot(color='red', ls='--', linewidth=1.5, label='UCL')
LCL.plot(color='red', ls='--', linewidth=1.5, label='LCL')
df_daily['MEAN'].plot(color='red', linewidth=2, label='Average')
plt.title('Total accidents per day', fontsize=16)
plt.xlabel('Day')
plt.ylabel('Number of accidents')
plt.tick_params(labelsize=14)
plt.legend(prop={'size':16})
```

```python
df['convert_to_date'] = pd.to_datetime(df['Date'])
```

```python
df['Day'] = df['convert_to_date'].dt.day
df['Month'] = df['convert_to_date'].dt.month
df['Year'] = df['convert_to_date'].dt.year
```

```python
df_pivot_table = df.pivot_table(values='Date', index='Day', columns='Month', aggfunc=len)
df_pivot_table_date_count = df.pivot_table(values='Date', index='Day', columns='Month', aggfunc=lambda x: len(x.unique()))
df_average = df_pivot_table/df_pivot_table_date_count
df_average.columns = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(7,9))
plt.title('Average number of accidents per day and month', fontsize=14)
sns.heatmap(df_average.round(), cmap='seismic', linecolor='grey',linewidths=0.1, cbar=False, annot=True, fmt=".0f")
```

```python
df_time_not_null = df[df['Time'] != 'N/A']

df_time_not_null['Hour'] = pd.to_datetime(df_time_not_null['Time'], format='%H:%M').dt.hour
```

```python
df_pivot_table = df_time_not_null.pivot_table(values='Date', index='Hour', columns='Day_of_Week', aggfunc=len)
df_pivot_table_date_count = df_time_not_null.pivot_table(values='Date', index='Hour', columns='Day_of_Week', aggfunc=lambda x: len(x.unique()))
df_average = df_pivot_table/df_pivot_table_date_count
df_average.columns = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']

plt.figure(figsize=(6,6))
plt.title('Average number of accidents per hour and day of the week', fontsize=14)
plt.tick_params(labelsize=12)
sns.heatmap(df_average.round(), cmap='seismic', linecolor='grey',linewidths=0.1, cbar=False, annot=True, fmt=".0f")
```

```python
plt.figure(figsize=(15,6))
df.resample('M').size().plot(label='Total per month')
df.resample('M').size().rolling(window=12).mean().plot(color='red', linewidth=5, label='12-months Moving Average')

plt.title('Accidents per month', fontsize=16)
plt.xlabel('')
plt.legend(prop={'size':16})
plt.tick_params(labelsize=16)
```
